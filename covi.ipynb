```python
import requests
import pandas as pd

# URL base de la API (datos diarios de EE.UU.)
url = "https://api.covidtracking.com/v1/us/daily.json"

# Petición a la API
response = requests.get(url)

# Verificamos que la respuesta sea correcta
if response.status_code == 200:
    data = response.json()  # Convertimos a lista de diccionarios
    df = pd.DataFrame(data)  # Convertimos a DataFrame para analizar
    print("Datos descargados correctamente ✅")
    print(df.head())  # Muestra las primeras filas
else:
    print("Error al acceder a la API:", response.status_code)
```
